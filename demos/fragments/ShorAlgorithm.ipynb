## Shor's Algorithm Demo

This is a demo notebook showing how Shor's algorithm is used to factor 15.

```python
from math import gcd
from qiskit import QuantumCircuit, QuantumRegister
import matplotlib.pyplot as plt
from qiskit.quantum_info import Statevector
import random

N = 15 # N is the number we want to factor
t = 11 # t is the number of qubits in the first register
print(f"N = {N} (number we want to factor)")
print(f"t = {t} (number of qubits in the first register)\n")

# {Classical} Step 1: find x in {2, ..., N-1} such that gcd(x, N) = 1
random.seed(42)
while True:
    x = random.randrange(2, N) # Sampling x until gcd(x,N)=1
    if gcd(x, N) == 1:
        break
print(f"x = {x}")
print(f"Check: gcd(x, N) = {gcd(x, N)}\n")

# {Classical} Step 2: find L such that N <= 2^L - 1, where L is the number of qubits in the second register.
# # Hardcode computation of L:
# L = 1
# while N > 2**L - 1:
#     L += 1
L = N.bit_length()
print(f"L = {L} (number of qubits in the second register)")
print(f"Check: 2^{L} - 1 = {2**L - 1} >= {N}\n")
```

```python
# {Quantum} Step 3, Quantum Phase Estimation: find k such that x^k = 1 (mod N).
reg1 = QuantumRegister(t, name="control")   # first register: t qubits (phase estimation)
reg2 = QuantumRegister(L, name="target")    # second register: L qubits (modular exponentiation)
qc = QuantumCircuit(reg1, reg2)
psi = Statevector(qc)
# psi.draw("latex")
for i in range(t):
    qc.h(i)

print(f"k = ")
qc.draw('mpl')
```
