```python

from LogicalQ import LogicalCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile
from qiskit.quantum_info import partial_trace, DensityMatrix

stabilizier_tableau_dict = {
    "phase_flip": {
        "label": (3,1,2),
        "tableau": [
            "XXI",
            "IXX",
        ],
    },
    "four": {
        "label": (4,2,2),
        "tableau": [
            "XZZX",
            "YZZY",
        ],
    },
    "five": {
        "label": (5,1,3),
        "tableau": [
            "XZZXI",
            "IXZZX",
            "XIXZZ",
            "ZXIXZ",
        ],
    },
    "steane": {
        "label": (7,1,3),
        "tableau": [
            "XXXXIII",
            "IXXIXXI",
            "IIXXIXX",
            "ZZZZIII",
            "IZZIZZI",
            "IIZZIZZ",
        ],
    },
    "eight": {
        "label": (8,3,3),
        "tableau": [
            "XXXXXXXX",
            "ZZZZZZZZ",
            "IXIXYZYZ",
            "IXZYIXZY",
            "IYXZXZIY",
        ],
    },
}
#TODO: find stabilizer tables for different stabilizer codes

def TestEncode():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        n, k, d = stabilizer_tableau['label']
        lqcirc_encode = LogicalCircuit(stabilizer_tableau)
        q_indices = list(range(k))
        lqcirc_encode.encode(q_indices)
        lqcirc_encode.measure_all()
        
def TestX():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        lqcirc_x = LogicalCircuit(stabilizer_tableau)
        n, k, d = stabilizer_tableau["label"]
        targets = list(range(k))
        lqcirc_x.x(targets)
        qc = lqcirc_x.circuit
        simulator = AerSimulator()
        tqc = transpile(qc, simulator)
        result = simulator.run(tqc).result()
        final_dm = result.data(0)['density_matrix']
        rho = DensityMatrix(final_dm)
        reduced = partial_trace(rho, list(range(k, n)))
        print(reduced.draw(output='text'))

def TestZ():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        lqcirc_z = LogicalCircuit(stabilizer_tableau)
        n, k, d = stabilizer_tableau["label"]
        targets = list(range(k))
        lqcirc_z.z(targets)
        lqcirc_z.measure_all()

def TestH():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        lqcirc_h = LogicalCircuit(stabilizer_tableau)
        n, k, d = stabilizer_tableau["label"]
        targets = list(range(k))
        lqcirc_h.h(targets)
        lqcirc_h.measure_all()

def TestY():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        lqcirc_y = LogicalCircuit(stabilizer_tableau)
        n, k, d = stabilizer_tableau["label"]
        targets = list(range(k))
        lqcirc_y.y(targets)
        lqcirc_y.measure_all()

def TestS():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        lqcirc_s = LogicalCircuit(stabilizer_tableau)
        n, k, d = stabilizer_tableau["label"]
        targets = list(range(k))
        lqcirc_s.s(targets)
        lqcirc_s.measure_all()

def TestT():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        lqcirc_t = LogicalCircuit(stabilizer_tableau)
        n, k, d = stabilizer_tableau["label"]
        targets = list(range(k))
        lqcirc_t.t(targets)
        lqcirc_t.measure_all()

def TestCX():
    for name, stabilizer_tableau in stabilizier_tableau_dict.items():
        n, k, d = stabilizer_tableau["label"]
        lqcirc_cx = LogicalCircuit(stabilizer_tableau["tableau"])
        control = 0
        targets = list(range(1, k)) 
        lqcirc_cx.cx(control, *targets)
        lqcirc_cx.measure_all()  
#TODO find expected results
#TODO density matrices
```
